```python
import os
import cv2
from pathomics_research_utils import utils
```

```python
lymphoma_dir = '/home/krypt/myStuff/projects/cancer-research/data/lymphoma'
lymphoma_to_dir = lymphoma_dir + '_2'
```

```python
def make_patches_folder(dirr, to_dirr):
    if not os.path.exists(to_dirr):
        os.mkdir(to_dirr)
    classes = os.listdir(dirr)
    print(classes)
    for cn in classes:
        if not os.path.exists(os.path.join(to_dirr, cn)):
            os.mkdir(os.path.join(to_dirr, cn))
    #generate 32x32 patches
    for cn in classes:
        cur = 0
        for fname in os.listdir(os.path.join(dirr, cn)):
            try:
                patches = utils.get_image_subpatches(utils.read_image(os.path.join(os.path.join(dirr,cn), fname)))
                for patch in patches:
                    cv2.imwrite(os.path.join(os.path.join(to_dirr,cn), str(cur)+".png"), patch)
                    cur += 1
            except:
                pass
```
